# Hardness Peak

Reproduces Table 2: hardness density H(n,α) = log T / n peaking at α*≈4.20.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.family': 'serif', 'font.size': 12})

from src.hardness_metrics import measure_hardness
from src.instance_generator import generate_instance_batch
from src.energy_model import barrier_density, ALPHA_D, ALPHA_S, ALPHA_STAR
```

```python

# Table 2 values (manuscript)
table2 = {100: 0.0789, 200: 0.0513, 400: 0.0336, 800: 0.0210}
print("n | H_manuscript | b(α*)/n·b(4.20)")
for n, H_ms in table2.items():
    print(f"  {n:4d} | {H_ms:.4f}       | {barrier_density(4.20):.4f}")
```

```python

alphas = np.linspace(3.4, 5.0, 34)
b_theory = np.array([barrier_density(a) for a in alphas])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4.5))

# Left: theoretical barrier
ax1.plot(alphas, b_theory, 'k-', lw=2.0)
ax1.axvline(ALPHA_STAR, color='red', ls='--', lw=1.0, label=f'α*={ALPHA_STAR}')
ax1.axvline(ALPHA_D,    color='gray', ls=':',  lw=0.9)
ax1.axvline(ALPHA_S,    color='gray', ls=':',  lw=0.9)
ax1.set_xlabel(r'α');  ax1.set_ylabel(r'b(α)');  ax1.set_title('Barrier density b(α)')
ax1.legend()

# Right: linear scaling log T ∝ n
ns = [100, 200, 400, 800]
log_T_theory = [n * barrier_density(4.20) for n in ns]
slope, intercept = np.polyfit(ns, log_T_theory, 1)
ax2.plot(ns, log_T_theory, 'ko', ms=6)
ax2.plot(ns, np.polyval([slope, intercept], ns), 'r--',
         label=f'slope={slope:.4f} (theory)')
ax2.set_xlabel('n');  ax2.set_ylabel('log T theory');  ax2.set_title('Barrier scaling')
ax2.legend()

plt.tight_layout()
plt.savefig('../results/figures/nb04_hardness_peak.png', dpi=150)
plt.show()
print(f"Theoretical slope at α*=4.20: {slope:.4f}  (manuscript: 0.021)")
```
