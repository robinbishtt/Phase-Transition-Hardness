```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.family': 'serif', 'font.size': 12})

from src.cryptography import KSATProofOfWork, pow_difficulty_parameter
from src.hardness_metrics import dpll_solve
```

```python

# Compute n needed for various security levels
for bits in [10, 20, 40, 60, 80]:
    n = pow_difficulty_parameter(bits)
    print(f"  {bits:3d} bits security → n = {n:6d} variables")
print()
```

```python

# Demonstrate puzzle generation and verification
pow_puzzle = KSATProofOfWork(n=20, alpha=4.20, k=3, master_seed=42)
puzzle = pow_puzzle.generate_puzzle(nonce=12345)
print(f"Puzzle: n={puzzle['n']}, α={puzzle['alpha']}, difficulty={puzzle['difficulty']:.4f}")
print(f"Challenge: {puzzle['challenge'][:32]}...")

# Attempt to solve (DPLL — easy for n=20)
result = dpll_solve(puzzle['instance'], max_decisions=100000)
if result['satisfiable']:
    valid = pow_puzzle.verify_solution(result['assignment'])
    print(f"Solution found: valid={valid}  ✓")
else:
    print("No solution found (UNSAT instance or cutoff)")
```
