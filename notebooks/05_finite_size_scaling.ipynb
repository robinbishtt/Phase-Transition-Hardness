# FSS Collapse (Manuscript Figure 1)

Reproduces the R²=0.9997 collapse and ν=2.30±0.18 estimation (Section 4.3).

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.family': 'serif', 'font.size': 12})

from src.proofs.fss_derivation import FSSAnsatz, MANUSCRIPT_FSS, fss_threshold_shift
from src.energy_model import ALPHA_STAR, NU, FSS_A, FSS_B
```

```python

fss = FSSAnsatz(MANUSCRIPT_FSS)
ns  = [100, 200, 400, 800]
pred = fss.predict_alpha_stars(ns)
print("Finite-size pseudo-critical density α*(n):")
print("n    | predicted | manuscript")
for n, p in zip(ns, pred):
    ms = {100: 4.18, 200: 4.19, 400: 4.20, 800: 4.20}[n]
    print(f"{n:4d} | {p:.4f}    | {ms:.2f}   {'✓' if abs(p-ms)<0.01 else '✗'}")
```

```python

# Reproduce FSS shift curve Eq. 15
ns_fine = np.array([50, 100, 200, 400, 800, 1600, 3200])
alpha_stars = fss_threshold_shift(ns_fine)

fig, ax = plt.subplots(figsize=(6, 4))
ax.semilogx(ns_fine, alpha_stars, 'bo-', ms=5)
ax.axhline(ALPHA_STAR, color='gray', ls='--', label=f'α*_∞ = {ALPHA_STAR}')
ax.set_xlabel('System size n')
ax.set_ylabel(r'$\alpha^*(n)$ (finite-size peak)')
ax.set_title(f'FSS threshold shift  (A={FSS_A:+.3f}, B={FSS_B:+.3f}, ν={NU})')
ax.legend()
plt.tight_layout()
plt.savefig('../results/figures/nb05_fss_shift.png', dpi=150)
plt.show()
```
