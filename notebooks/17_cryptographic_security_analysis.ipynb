```python

# PRG parameters from Section 5.2
for eps in [0.010, 0.020, 0.025]:
    prg = APKPseudoRandomGenerator(n=400, epsilon=eps, alpha=4.20)
    print(f"ε={eps:.3f}: output={prg.output_length()} bits, "
          f"security={prg.security_level():.1f} bits, "
          f"ε < Σ(4.20)={prg.sigma:.4f}: {eps < prg.sigma}")

print()
# Full PRG report for the manuscript example (ε just below Σ(4.20) ≈ 0.027)
prg_ms = APKPseudoRandomGenerator(n=400, epsilon=0.025, alpha=4.20)
report = prg_ms.prg_parameters()
print("PRG parameter report (manuscript Section 5.2):")
print(f"  Seed length:  {report['seed_length_n']} bits")
print(f"  Output length:{report['output_length']} bits")
print(f"  Stretch ε:    {report['stretch_epsilon']:.3f} (< Σ={report['sigma_alpha']:.4f})")
print(f"  Security:     {report['security_bits']:.1f} bits")
print("  AIK conditions:")
for cname, cond in report['aik_conditions'].items():
    print(f"    {cname:30s}: {cond['satisfied']}  — {cond['reason'][:60]}")
```
