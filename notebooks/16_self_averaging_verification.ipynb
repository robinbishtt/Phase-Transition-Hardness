```python

# Reproduce Table 5 structure from manuscript
# The paper shows: σ_H/E[H] decreases but slower than n^{-1/2} due to heavy tail
print("Self-averaging check (manuscript Table 5 values):")
print(f"{'n':>5} | {'σ_logT/n':>10} | {'σ_H/E[H]':>10} | {'n^-0.5':>8} | {'slower?'}")
ms_data = {100:(0.0234,0.28), 200:(0.0228,0.19), 400:(0.0223,0.13), 800:(0.0196,0.09)}
prev_cv = None
for n, (sigma, cv) in ms_data.items():
    predicted = n**-0.5
    slower = (prev_cv is None) or (cv > predicted)
    print(f"{n:5d} | {sigma:.4f}     | {cv:.4f}     | {predicted:.4f}   | {slower}")
    prev_cv = cv
print("\n→ CV decreases slower than n^{-1/2} due to non-Gaussian tail of runtime distribution")
```
