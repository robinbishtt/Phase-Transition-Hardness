# Intensive Barrier Function (Manuscript Table 4)

Validates the calibrated b(α) formula and critical scaling b ~ (α-α_d)^κ.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.family': 'serif', 'font.size': 12})

from src.energy_model import barrier_density, ALPHA_D, ALPHA_S, KAPPA, NU, ETA
```

```python

# Table 4 values
table4 = {3.5: 0.003, 3.8: 0.012, 4.0: 0.020, 4.2: 0.021, 4.4: 0.015, 4.6: 0.008}
print("Table 4 verification:")
print(f"{'α':>5} | {'b(α) code':>10} | {'b(α) MS':>8} | {'Regime'}")
for alpha, b_ms in table4.items():
    b_code = barrier_density(alpha)
    regime = "UNSAT†" if alpha > ALPHA_S else ("Peak" if alpha == 4.2 else "Hard-SAT")
    print(f"{alpha:5.1f} | {b_code:10.4f} | {b_ms:8.3f} | {regime}")
```

```python

# Critical scaling b(α) ~ (α - α_d)^κ
alphas_near = np.linspace(ALPHA_D + 0.02, 4.15, 50)
b_vals = np.array([barrier_density(a) for a in alphas_near])
b_vals_crit = 0.031 * (alphas_near - ALPHA_D) ** KAPPA

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4))
ax1.loglog(alphas_near - ALPHA_D, b_vals, 'bo', ms=4, label='b(α) code')
ax1.loglog(alphas_near - ALPHA_D, b_vals_crit, 'r--', label=f'0.031·(α−α_d)^{KAPPA}')
ax1.set_xlabel(r'α − α_d');  ax1.set_ylabel(r'b(α)')
ax1.set_title(f'Critical scaling κ = ν(1−η) = {KAPPA:.2f}')
ax1.legend()

alphas_all = np.linspace(ALPHA_D + 0.01, ALPHA_S - 0.01, 200)
ax2.plot(alphas_all, [barrier_density(a) for a in alphas_all], 'k-', lw=2)
ax2.axvline(4.20, color='red', ls='--', label='α*=4.20')
ax2.set_xlabel(r'α');  ax2.set_ylabel(r'b(α)');  ax2.set_title('Barrier density b(α)')
ax2.legend()
plt.tight_layout()
plt.savefig('../results/figures/nb14_barrier_function.png', dpi=150)
plt.show()
print(f"Mean field prediction κ_MF = ν = {NU:.2f} is EXCLUDED at >5σ")
print(f"Measured:  κ = {KAPPA:.2f} (= ν·(1−η) = {NU}×{1-ETA:.2f})")
```
