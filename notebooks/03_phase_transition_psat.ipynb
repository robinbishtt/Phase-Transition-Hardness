# Satisfiability Probability P_sat(α, n)

Reproduces the P_sat curves and threshold location (Section 4.1).

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.family': 'serif', 'font.size': 12})

from src.phase_transition import psat_curve, locate_threshold, theoretical_order_parameters
from src.energy_model import ALPHA_S
```

```python

alphas = np.linspace(3.5, 5.0, 30)
ns     = [30, 50, 80]   # small n for notebook speed
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

fig, ax = plt.subplots(figsize=(7, 4.5))
for n, col in zip(ns, colors):
    psats = psat_curve(n=n, alphas=alphas, n_instances=100, k=3, master_seed=20240223,
                       solver='dpll', n_jobs=1)
    alpha_s_est = locate_threshold(alphas, psats, target=0.5)
    ax.plot(alphas, psats, color=col, lw=1.8, label=f'n={n}  (α_s≈{alpha_s_est:.3f})')

ax.axvline(ALPHA_S, color='gray', ls='--', lw=1.0, label=f'α_s={ALPHA_S} (lit.)')
ax.axhline(0.5, color='gray', lw=0.7, ls=':')
ax.set_xlabel(r'Clause density $\alpha$')
ax.set_ylabel(r'$P_{\rm sat}(\alpha, n)$')
ax.set_title('Satisfiability phase transition in random 3-SAT')
ax.legend(fontsize=9)
ax.set_ylim(-0.02, 1.08)
plt.tight_layout()
plt.savefig('../results/figures/nb03_psat.png', dpi=150)
plt.show()
```
