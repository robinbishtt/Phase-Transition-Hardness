# Critical Exponent ν = 2.30 ± 0.18

Reproduces Table 3: three independent estimators converge to ν=2.30.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.family': 'serif', 'font.size': 12})

from src.binder_cumulant import BinderCumulant, CriticalExponentEstimator, compute_binder_crossing
```

```python

ce = CriticalExponentEstimator(ns=[100, 200, 400, 800])

r_peak = ce.nu_from_peak_shift()
r_binder = ce.nu_from_binder()
r_combined = ce.combined_estimate()

print("Reproduction of Table 3:")
print(f"  Peak-shift:     ν = {r_peak['nu']:.2f}  CI {r_peak['ci']}")
print(f"  Binder:         ν = {r_binder['nu']:.2f}  CI {r_binder['ci']}")
print(f"  Combined:       ν = {r_combined['nu']:.2f}  CI {r_combined['ci']}")
print(f"  Cavity (lit.):  ν = 2.35  CI [2.30, 2.40]")
print(f"  σ from cavity: {r_combined['sigma_from_cavity']:.2f}σ  (manuscript: 0.28σ)")
```

```python

bc = BinderCumulant(ns=[100, 200, 400, 800])
alphas = np.linspace(4.0, 4.5, 80)
curves = bc.binder_curves(alphas)

fig, ax = plt.subplots(figsize=(6, 4))
for n, U in sorted(curves.items()):
    ax.plot(alphas, U, lw=1.8, label=f'n={n}')
val = bc.validate_crossing(alphas)
ax.axvline(val['alpha_crossing'], color='gray', ls='--',
           label=f"crossing α_c={val['alpha_crossing']:.3f}")
ax.set_xlabel(r'α');  ax.set_ylabel(r'$U_n(\alpha)$')
ax.set_title(f"Binder cumulant crossing (α_c={val['alpha_crossing']:.3f} ± {bc.ALPHA_C_UNCERTAINTY})")
ax.legend(fontsize=9)
plt.tight_layout()
plt.savefig('../results/figures/nb06_binder.png', dpi=150)
plt.show()
```
