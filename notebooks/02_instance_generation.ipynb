# Instance Generation

Verifies deterministic generation, clause properties, and SHA-256 seeding.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.family': 'serif', 'font.size': 12})

from src.instance_generator import (
    generate_ksat_instance, generate_instance_batch,
    count_violated_clauses, is_satisfying,
)
from src.utils import derive_seed
```

```python

# Paper master seed
MASTER = 20240223

# Verify SHA-256 seeding is deterministic across Python versions
s1 = derive_seed(MASTER, 100, 4.2, 0)
s2 = derive_seed(MASTER, 100, 4.2, 0)
assert s1 == s2, "Seeds must be identical!"
print(f"derive_seed(MASTER, 100, 4.2, 0) = {s1}  (deterministic ✓)")

inst = generate_ksat_instance(50, 4.2, k=3, seed=s1)
print(f"Instance: n={inst['n']}, m={inst['m']}, k={inst['k']}, α={inst['alpha']}")
print(f"Each clause has {len(inst['clauses'][0])} literals  ✓")
for clause in inst['clauses']:
    assert len(set(abs(l) for l in clause)) == 3, "Duplicate variables in clause!"
print("No duplicate variables in any clause  ✓")
```

```python

# Reproduce paper batch: n=100, α=4.20, 10 instances, master seed
batch = generate_instance_batch(100, 4.20, 10, k=3, master_seed=MASTER)
assert len(batch) == 10
# All instances must differ
clauses_set = set(tuple(tuple(c) for c in b['clauses']) for b in batch)
assert len(clauses_set) == 10, "All 10 instances must be distinct"
print(f"Generated {len(batch)} distinct instances  ✓")
```
