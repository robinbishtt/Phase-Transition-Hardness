# Conjecture 4: log T(n,α) = Θ(n·b(α))

Verifies the sandwich bounds and regression R²=0.9876 from Section 4.2.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.family': 'serif', 'font.size': 12})

from src.proofs.runtime_bounds import RuntimeBounds, conjecture4_bounds, verify_theta_scaling
from src.energy_model import barrier_density, ALPHA_D, ALPHA_S
```

```python

rb = RuntimeBounds()
ns = [100, 200, 400, 800]
report = rb.evaluate_grid(ns, np.linspace(ALPHA_D+0.05, ALPHA_S-0.05, 40))
print(f"Mean sandwich ratio (upper/lower): {report.mean_sandwich_ratio:.2f}")
print(f"Conjecture-consistent fraction:    {report.conjecture_support:.3f}")
print(f"Critical exponent ν:               {report.critical_exponent_nu}")
print(f"Barrier growth exponent κ:          {report.barrier_growth_exponent_kappa}")
```

```python

# Verify Θ scaling at peak hardness
v = verify_theta_scaling(ns=[100,200,400,800], alpha=4.20, regression_slope=0.0122)
print("\nTheta scaling verification at α=4.20:")
for k, val in v.items():
    print(f"  {k}: {val}")
```

```python

alphas = np.linspace(ALPHA_D + 0.05, ALPHA_S - 0.05, 60)
lb_curve = rb.lower.lower_bound_curve(400, alphas)
ub_curve = rb.upper.upper_bound_curve(400, alphas)
theory   = np.array([400 * barrier_density(a) for a in alphas])

fig, ax = plt.subplots(figsize=(7, 4.5))
ax.fill_between(alphas, lb_curve, ub_curve, alpha=0.2, color='blue', label='Sandwich [lb, ub]')
ax.plot(alphas, theory, 'k-', lw=2.0, label=r'n·b(α) (theory)')
ax.set_xlabel(r'α');  ax.set_ylabel(r'log T bound (n=400)')
ax.set_title('Conjecture 4: Arrhenius lower + conflict-graph upper bound')
ax.legend()
plt.tight_layout()
plt.savefig('../results/figures/nb07_conjecture4.png', dpi=150)
plt.show()
```
