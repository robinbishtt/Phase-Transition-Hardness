## Phase Diagram of Random 3-SAT

```python

alphas = np.linspace(2.5, 4.8, 300)
entropy    = [rs_entropy_density(a) / np.log(2) for a in alphas]
complexity = [cluster_complexity(a) * 2 for a in alphas]
frozen     = [frozen_fraction(a) for a in alphas]
barrier    = [barrier_density(a) / 0.021 for a in alphas]

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.plot(alphas, entropy,    label=r'Entropy $s/\log 2$')
ax.plot(alphas, complexity, label=r'Complexity $2\Sigma$')
ax.plot(alphas, frozen,     label=r'Frozen fraction')
ax.plot(alphas, barrier,    label=r'Barrier $b/b_{max}$')
for v, l in [(ALPHA_D,'α_d'),(ALPHA_S,'α_s')]:
    ax.axvline(v, color='gray', ls='--', lw=0.9)
    ax.text(v+0.02, 1.04, l, transform=ax.get_xaxis_transform(), fontsize=10, color='gray')
ax.axvspan(ALPHA_D, ALPHA_S, alpha=0.07, color='red', label='Hard phase')
ax.set_xlabel(r'Clause density $\alpha$')
ax.legend(fontsize=9)
ax.set_ylim(-0.05, 1.15)
plt.tight_layout()
plt.savefig('../results/figures/nb01_phase_diagram.png', dpi=150)
plt.show()
print("Figure saved.")
```
